```python
import matplotlib.pyplot as plt
from awpy.visualization.plot import plot_map
```

```python
maps = ['de_ancient','de_dust2', 'de_inferno', 'de_mirage', 'de_nuke', 'de_overpass', 'de_vertigo']

for m in maps:
    f, a = plot_map(map_name=m, map_type="simple", dark=True)
    
    f.set_size_inches(9, 9)
    plt.title(m)
    plt.savefig("radar_images/" + m + ".png")
    plt.show()
```
